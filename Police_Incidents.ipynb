```python
import pandas as pd
import numpy as np
import folium

df=pd.read_csv('Police_Incidents_2016.csv')
df.head()
```

```python
df.shape
```

```python
df=df.iloc[0:100,:]
df.shape
```

```python

sanfran_map = folium.Map(location=[37.77,-122.42], zoom_start=12)
sanfran_map
```

```python
incidents=folium.map.FeatureGroup()

for lat,lng in zip(df.Y,df.X):
    incidents.add_child(
        folium.vector_layers.CircleMarker(
            [lat,lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    ).add_to(sanfran_map)
    
lats=list(df.Y)
longs=list(df.X)
labels=list(df.Category)

for lat, lng, label in zip(lats, longs, labels):
    folium.Marker([lat, lng], popup=label).add_to(sanfran_map)  

sanfran_map.add_child(incidents)

sanfran_map
```
